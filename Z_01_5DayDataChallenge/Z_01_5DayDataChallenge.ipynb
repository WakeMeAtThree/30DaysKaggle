# Day 3

```python
import pandas as pd
import matplotlib.pyplot as plt 
import pylab
from scipy.stats import ttest_ind, probplot # t-test, qqplot from scipy.stats
```

```python
cereals = pd.read_csv("../../datasets/cereal.csv")
cereals.head()
```

```python
# plot a qqplot to check normality. If variable is normally distributed, most of the points should be along
# the center diagonal 
probplot(cereals["sodium"], dist="norm", plot=pylab)
```

```python
hotCereals = cereals["sodium"][cereals["type"]== "H"]
coldCereals = cereals["sodium"][cereals["type"]== "C"]

ttest_ind(hotCereals, coldCereals, equal_var = False)
```

```python
print("Mean sodium for the hot cereals:")
print(hotCereals.mean())

print("Mean sodium for the cold cereals:")
print(coldCereals.mean())
```

```python
plt.hist(coldCereals, alpha=0.5, label='cold')
plt.hist(hotCereals, label='hot')
plt.legend(loc = 'upper right')
plt.title("Sodium(mg) content of cereals by type")
```

# Day 4

```python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_csv("../../datasets/anonymous-survey-responses.csv")
dataframe.head()
```

```python
petFreqTable = dataframe['Just for fun, do you prefer dogs or cat?'].value_counts()

list(petFreqTable.index)
petFreqTable.values

labels = list(petFreqTable.index)
positionsForBars = list(range(len(labels)))

plt.bar(positionsForBars, petFreqTable.values)
plt.xticks(positionsForBars, labels)
plt.title("Pet Preferences")
```

```python
import seaborn as sns

sns.countplot(dataframe["Just for fun, do you prefer dogs or cat?"]).set_title("Dogs vs. Cats")
```

# Day 5

```python
import scipy.stats
import pandas as pd
```

```python
surveyData = pd.read_csv("../../datasets/anonymous-survey-responses.csv")
surveyData.head()
```

```python
scipy.stats.chisquare(surveyData["Have you ever taken a course in statistics?"].value_counts())
```

```python
scipy.stats.chisquare(surveyData["Do you have any previous experience with programming?"].value_counts())
```

```python
contingencyTable = pd.crosstab(surveyData["Do you have any previous experience with programming?"],
                               surveyData["Have you ever taken a course in statistics?"])
scipy.stats.chi2_contingency(contingencyTable)
```
